# Importaciones según principales países de origen

```python
# Agrupar por la columna "DESPAI" y sumamos los valores "FRO"
df_despais_suma = im[["DESPAI","FRO"]].groupby(by=["DESPAI"]).sum().reset_index()

# Realizamos un ordenamiento de los paises y sacamos el "top 10"
df_top10_despais = df_despais_suma.sort_values(by='FRO', ascending=False).head(10)

# Creamos una nueva fila para el "resto" y la añadimos
df_top10_despais = df_top10_despais.append({'DESPAI': 'Otros Paises',
                          'FRO': df_despais_suma.loc[~df_despais_suma['DESPAI'].isin(df_top10_despais['DESPAI']),'FRO'].sum()}, ignore_index=True) #~=not

# Calcular la columna de porcentaje
total_valor_despais = df_despais_suma['FRO'].sum()
df_top10_despais['Porcentaje'] = (df_top10_despais['FRO'] / total_valor_despais) * 100

# Creamos la fila de total
total_fila_despais = {'DESPAI': 'Total',
             'FRO': total_valor_despais,
             'Porcentaje': df_top10_despais["Porcentaje"].sum()}
df_top10_despais = df_top10_despais.append(total_fila_despais, ignore_index=True) #añadimos la fila

# Colocamos de mayor a menor
df_top10_despais
```

# Exportaciones según modo de transporte por Volumen

```python
# Agrupar por la columna "DESMED" y sumar el volumen "KILOS"
df_desmedi_suma = im[["DESMED","KILOS"]].groupby(by=["DESMED"]).sum().reset_index().sort_values("KILOS", ascending=False) #filtramos en una variable

# Calcular el total y el porcentaje
total_valor_desmedi = df_desmedi_suma["KILOS"].sum() #obtenemos el total
df_desmedi_suma['Porcentaje'] = df_desmedi_suma["KILOS"] / total_valor_desmedi * 100 #creamos columna de porcentaje

# Creamos la fila de total a partir de un diccionario
total_fila_desmedi = {'DESMED': 'Total',
             "KILOS": total_valor_desmedi,
             'Porcentaje': df_desmedi_suma["Porcentaje"].sum()}
df_desmedi_suma = df_desmedi_suma.append(total_fila_desmedi, ignore_index=True)

df_desmedi_suma
```

# Importaciones según Departamento de la Aduana de ingreso

```python
# Agrupar por la columna "DESDEPTO" y "FRO"
df_desdep_suma = im[["DESDEPTO","FRO",]].groupby(by=["DESDEPTO"]).sum().reset_index().sort_values("FRO", ascending=False) #filtramos en una variable

# Calcular el total y el porcentaje
total_valor_desdep = df_desdep_suma['FRO'].sum() #obtenemos el total
df_desdep_suma['Porcentaje'] = df_desdep_suma['FRO'] / total_valor_desdep * 100 #creamos columna de porcentaje

# Creamos la fila de total a partir de un diccionario
total_fila_desdep = {'DESDEPTO': 'Total',
             'FRO': total_valor_desdep,
             'Porcentaje': df_desdep_suma["Porcentaje"].sum()}
df_desdep_suma = df_desdep_suma.append(total_fila_desdep, ignore_index=True)

df_desdep_suma
```

# Top 10 de las importaciones bolivianas

```python
# 1. Obtenemos el top 10 de Importaciones
top_productos_im = im.groupby('DESNAN')['FRO'].sum().nlargest(10).index.tolist()

# 2. Inicializar una lista para almacenar los resultados finales
resultados = []

# 3. Iterar sobre cada producto en el top 10
for producto in top_productos_im:
    df_producto = im[im['DESNAN'] == producto]

    # Obtener el top 5 de País de Exportación según Valor
    top_paises_im = df_producto.groupby('DESPAI').agg({'FRO': 'sum', 'KILOS': 'sum'}).nlargest(5, 'FRO').reset_index()

    # Agregar una fila adicional para otros países
    otros_paises_valor_tr = df_producto[~df_producto['DESPAI'].isin(top_paises_im['DESPAI'])]['FRO'].sum()
    otros_paises_volumen_tr = df_producto[~df_producto['DESPAI'].isin(top_paises_im['DESPAI'])]['KILOS'].sum()

    otros_paises_tr = pd.DataFrame({
        'DESPAI': ['Otros Paises'],
        'FRO': [otros_paises_valor_tr],
        'KILOS': [otros_paises_volumen_tr]
    })

    #Concatenamos la fila de otros paises
    top_paises_im = pd.concat([top_paises_im, otros_paises_tr], ignore_index=True)

    # Calcular el porcentaje del Valor
    top_paises_im['Porcentaje_Valor'] = (top_paises_im['FRO'] / df_producto['FRO'].sum()) * 100

    # Agregar una fila de total para cada producto
    total_producto = df_producto.groupby('DESNAN').agg({'FRO': 'sum', 'KILOS': 'sum'}).reset_index()
    total_producto['DESPAI'] = 'Total Importado'
    total_producto['Porcentaje_Valor'] = 100  # 100% para la fila total
    top_paises_im = pd.concat([top_paises_im, total_producto], ignore_index=True)

    # Agregar el Producto al DataFrame final
    top_paises_im['DESNAN'] = producto

    # Agregar los resultados al listado final
    resultados.append(top_paises_im)

# Concatenar los resultados en un solo DataFrame
df_im_pruducto_final = pd.concat(resultados, ignore_index=True)

# Cambiamos de lugar algunas columnas
df_im_pruducto_final = df_im_pruducto_final[["DESNAN", "DESPAI", "KILOS", "FRO", "Porcentaje_Valor"]]
#Imprimimos el resultado final
df_im_pruducto_final.head(14)
```

# Importaciones según zonas geograficas

```python
# 1. Obtenemos las 3 zonas geograficas
top_im_zonas_geo = im.groupby('OTROS')['FRO'].sum().nlargest(4).index.tolist()

# 2. Inicializar una lista para almacenar los resultados finales
resultados_zonas_geo = []

# 3. Iterar sobre cada zona
for zona in top_im_zonas_geo:
    df_zona = im[im['OTROS'] == zona]

    # Obtener el top 10 de productos que mas se exportan
    top_productos_im_zona_geo = df_zona.groupby('DESNAN').agg({'FRO': 'sum', 'KILOS': 'sum'}).nlargest(10, 'FRO').reset_index()

    # Agregar una fila adicional para otros productos
    otros_productos_valor_g = df_zona[~df_zona['DESNAN'].isin(top_productos_im_zona_geo['DESNAN'])]['FRO'].sum()
    otros_productos_volumen_g = df_zona[~df_zona['DESNAN'].isin(top_productos_im_zona_geo['DESNAN'])]['KILOS'].sum()

    otros_productos_g = pd.DataFrame({
        'DESNAN': ['Otros Productos'],
        'FRO': [otros_productos_valor_g],
        'KILOS': [otros_productos_volumen_g]
    })

    #Concatenamos la fila de otros productos
    top_productos_im_zona_geo = pd.concat([top_productos_im_zona_geo, otros_productos_g], ignore_index=True)

    # Calcular el porcentaje del Valor
    top_productos_im_zona_geo['Porcentaje_Valor'] = (top_productos_im_zona_geo['FRO'] / df_zona['FRO'].sum()) * 100

    # Agregar una fila de total para cada zona geografica
    total_producto_z = df_zona.groupby('OTROS').agg({'FRO': 'sum', 'KILOS': 'sum'}).reset_index()
    total_producto_z['DESNAN'] = f'Total exportado ({len(df_zona["DESNAN"].unique())} productos)'
    total_producto_z['Porcentaje_Valor'] = 100  # 100% para la fila total
    top_productos_im_zona_geo = pd.concat([top_productos_im_zona_geo, total_producto_z], ignore_index=True)

    # Agregar el zona al DataFrame final
    top_productos_im_zona_geo['OTROS'] = zona

    # Agregar los resultados al listado final
    resultados_zonas_geo.append(top_productos_im_zona_geo)

# Concatenar los resultados en un solo DataFrame
df_im_zonas_geo_final = pd.concat(resultados_zonas_geo, ignore_index=True)

# Cambiamos de lugar algunas columnas
df_im_zonas_geo_final = df_im_zonas_geo_final[["OTROS", "DESNAN", "KILOS", "FRO", "Porcentaje_Valor"]]
#Imprimimos el resultado final
df_im_zonas_geo_final.head(14)
```

# Importaciones segun principales paises de destino

```python
# 1. Obtenemos el top 5 de los paises a los que mas se exporta
top_im_paises_destino = im.groupby('DESPAI')['FRO'].sum().nlargest(5).index.tolist()

# 2. Inicializar una lista para almacenar los resultados finales
resultados_im_destino = []

# 3. Iterar sobre cada pais en el top 5
for pais in top_im_paises_destino:
    df_pais = im[im['DESPAI'] == pais]

    # Obtener el top 10 de de productos exportados segun pais
    top_productos_im_destino = df_pais.groupby('DESNAN').agg({'FRO': 'sum', 'KILOS': 'sum'}).nlargest(10, 'FRO').reset_index()

    # Agregar una fila adicional para otros productos
    otros_productos_valor_dt = df_pais[~df_pais['DESNAN'].isin(top_productos_im_destino['DESNAN'])]['FRO'].sum()
    otros_productos_volumen_dt = df_pais[~df_pais['DESNAN'].isin(top_productos_im_destino['DESNAN'])]['KILOS'].sum()

    otros_productos_dt = pd.DataFrame({
        'DESNAN': ['Resto de Productos'],
        'FRO': [otros_productos_valor_dt],
        'KILOS': [otros_productos_volumen_dt]
    })

    #Concatenamos la fila de otros productos
    top_productos_im_destino = pd.concat([top_productos_im_destino, otros_productos_dt], ignore_index=True)

    # Calcular el porcentaje del Valor
    top_productos_im_destino['Porcentaje_Valor'] = (top_productos_im_destino['FRO'] / df_pais['FRO'].sum()) * 100

    # Agregar una fila de total para cada pais
    total_pais = df_pais.groupby('DESPAI').agg({'FRO': 'sum', 'KILOS': 'sum'}).reset_index()
    total_pais['DESNAN'] = f'Total exportado ({len(df_pais["DESNAN"].unique())} productos)'
    total_pais['Porcentaje_Valor'] = 100  # 100% para la fila total
    top_productos_im_destino = pd.concat([top_productos_im_destino, total_pais], ignore_index=True)

    # Agregar el PAIS al DataFrame final
    top_productos_im_destino['DESPAI'] = pais

    # Agregar los resultados al listado final
    resultados_im_destino.append(top_productos_im_destino)

# Concatenar los resultados en un solo DataFrame
df_im_destino_final = pd.concat(resultados_im_destino, ignore_index=True)

# Cambiamos de lugar algunas columnas
df_im_destino_final = df_im_destino_final[["DESPAI", "DESNAN", "KILOS", "FRO", "Porcentaje_Valor"]]
#Imprimimos el resultado final
df_im_destino_final.head(12)
```

# Importaciones según departamentos

```python
# 1. Obtenemos los departamentos y la cantidad de exportaciones
top_im_departamentos = im.groupby('DESDEPTO')['FRO'].sum().nlargest(9).index.tolist()

# 2. Inicializar una lista para almacenar los resultados finales
resultados_im_departamento = []

# 3. Iterar sobre cada departamento (usamos el DF original)
for departamento in top_im_departamentos:
    df_departamento = im[im['DESDEPTO'] == departamento]

    # Obtener el top 10 de de productos exportados segun departamento
    top_productos_im_departamento = df_departamento.groupby('DESNAN').agg({'FRO': 'sum', 'KILOS': 'sum'}).nlargest(10, 'FRO').reset_index()

    # Agregar una fila adicional para otros productos
    otros_productos_valor_dp = df_departamento[~df_departamento['DESNAN'].isin(top_productos_im_departamento['DESNAN'])]['FRO'].sum()
    otros_productos_volumen_dp = df_departamento[~df_departamento['DESNAN'].isin(top_productos_im_departamento['DESNAN'])]['KILOS'].sum()

    otros_productos_dp = pd.DataFrame({
        'DESNAN': ['Resto de Productos'],
        'FRO': [otros_productos_valor_dp],
        'KILOS': [otros_productos_volumen_dp]
    })

    #Concatenamos la fila de otros productos
    top_productos_im_departamento = pd.concat([top_productos_im_departamento, otros_productos_dp], ignore_index=True)

    # Calcular el porcentaje del Valor
    top_productos_im_departamento['Porcentaje_Valor'] = (top_productos_im_departamento['FRO'] / df_departamento['FRO'].sum()) * 100

    # Agregar una fila de total para cada departamento
    total_departamento = df_departamento.groupby('DESDEPTO').agg({'FRO': 'sum', 'KILOS': 'sum'}).reset_index()
    total_departamento['DESNAN'] = f'Total importado ({len(df_departamento["DESNAN"].unique())} productos)'
    total_departamento['Porcentaje_Valor'] = 100  # 100% para la fila total
    top_productos_im_departamento = pd.concat([top_productos_im_departamento, total_departamento], ignore_index=True)

    # Agregar el departamento al DataFrame final
    top_productos_im_departamento['DESDEPTO'] = departamento

    # Agregar los resultados al listado final
    resultados_im_departamento.append(top_productos_im_departamento)

# Concatenar los resultados en un solo DataFrame
df_im_departamento_final = pd.concat(resultados_im_departamento, ignore_index=True)

# Cambiamos de lugar algunas columnas
df_im_departamento_final = df_im_departamento_final[["DESDEPTO", "DESNAN", "KILOS", "FRO", "Porcentaje_Valor"]]
#Imprimimos el resultado final
df_im_departamento_final.head(12)
```

# Importaciones según vias de ingreso

```python
# 1. Obtenemos las 13 vias de ingreso
top_im_via_salida = im.groupby('DESVIA')['FRO'].sum().nlargest(13).index.tolist()

# 2. Inicializar una lista para almacenar los resultados finales
resultados_im_via_salida = []

# 3. Iterar sobre cada via salida (usamos el DF original)
for via_salida in top_im_via_salida:
    df_via_salida = im[im['DESVIA'] == via_salida]

    # Obtener el top 5 de productos importados segun via de salida
    top_productos_im_via_salida = df_via_salida.groupby('DESNAN').agg({'FRO': 'sum', 'KILOS': 'sum'}).nlargest(5, 'FRO').reset_index()

    # Agregar una fila adicional para otros productos
    otros_productos_valor_vs = df_via_salida[~df_via_salida['DESNAN'].isin(top_productos_im_via_salida['DESNAN'])]['FRO'].sum()
    otros_productos_volumen_vs = df_via_salida[~df_via_salida['DESNAN'].isin(top_productos_im_via_salida['DESNAN'])]['KILOS'].sum()

    otros_productos_vs = pd.DataFrame({
        'DESNAN': ['Resto de Productos'],
        'FRO': [otros_productos_valor_vs],
        'KILOS': [otros_productos_volumen_vs]
    })

    #Concatenamos la fila de otros productos
    top_productos_im_via_salida = pd.concat([top_productos_im_via_salida, otros_productos_vs], ignore_index=True)

    # Calcular el porcentaje del Valor
    top_productos_im_via_salida['Porcentaje_Valor'] = (top_productos_im_via_salida['FRO'] / df_via_salida['FRO'].sum()) * 100

    # Agregar una fila de total para cada via de salida
    total_via_salida = df_via_salida.groupby('DESVIA').agg({'FRO': 'sum', 'KILOS': 'sum'}).reset_index()
    total_via_salida['DESNAN'] = f'Total importado ({len(df_via_salida["DESNAN"].unique())} productos)'
    total_via_salida['Porcentaje_Valor'] = 100  # 100% para la fila total
    top_productos_im_via_salida = pd.concat([top_productos_im_via_salida, total_via_salida], ignore_index=True)

    # Agregar la via de salida al DataFrame final
    top_productos_im_via_salida['DESVIA'] = via_salida

    # Agregar los resultados al listado final
    resultados_im_via_salida.append(top_productos_im_via_salida)

# Concatenar los resultados en un solo DataFrame
df_im_via_salida_final = pd.concat(resultados_im_via_salida, ignore_index=True)

# Cambiamos de lugar algunas columnas
df_im_via_salida_final = df_im_via_salida_final[['DESVIA', "DESNAN", "KILOS", "FRO", "Porcentaje_Valor"]]
#Imprimimos el resultado final
df_im_via_salida_final.head(7)
```
